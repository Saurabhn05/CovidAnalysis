```python
df_time = df.groupby('ObservationDate').agg({
    'Confirmed': 'sum',
    'Deaths': 'sum',
    'Recovered': 'sum'})
df_time
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(12,4))
plt.title('Time series of Confirmed Deaths and Recovered')
plt.xlabel('Obervation Date')
plt.ylabel('Count')
plt.legend()

sns.lineplot(data=df_time, x='ObservationDate', y='Confirmed', label='Confirmed')
sns.lineplot(data=df_time, x='ObservationDate', y='Deaths',  label='Deaths')
sns.lineplot(data=df_time, x='ObservationDate', y='Recovered',  label='Recovered')
plt.show()
```

```python
df_country = df.groupby('Country/Region').agg({
    'Confirmed': 'sum',
    'Deaths': 'sum',
    'Recovered': 'sum'})
```

```python
plt.figure(figsize=(8,4))
sns.barplot(x=df_country.index, y='Confirmed', data=df_country)
plt.xlabel('Country / Region')
plt.ylabel('Confirmed Case')
plt.title('Confirmed Cases By Country/Region')
plt.show()
```

```python
correct_analyst=df[['Confirmed','Deaths','Recovered']].corr()
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(correct_analyst,cmap='viridis',annot=True)
plt.title('Heatmap for Confirmed ,Deaths And Recovered Cases')
plt.show()
```

```python
df_country = df.groupby('Country/Region').agg({
    'Confirmed': 'sum',
    'Deaths': 'sum',
    'Recovered': 'sum'
})
df_country
```

```python
highest_confirmed=df_country.sort_values(by='Confirmed',ascending=False)
print('Highest Countries/Regions By Confirmed Cases:')
print(highest_confirmed.head())
```

```python
highest_deaths=df_country.sort_values(by='Deaths',ascending=False)
print('Highest Countries/Regions By Death Cases:')
print(highest_deaths.head())
```

```python
highest_recovered=df_country.sort_values(by='Recovered',ascending=False)
print('Highest Countries/Regions By Recovered Cases:')
print(highest_recovered.head())
```
